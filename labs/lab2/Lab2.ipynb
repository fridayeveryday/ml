```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# %matplotlib notebook
```

$y = sin(x_1 + x_1)$

## 1

```python
x_1 = np.linspace(0, 100, num = 500)
x_2 = np.linspace(-50, 80, num = 500)
```

```python
def sin_sum(x_1, x_2):
    return np.sin(x_1 + x_2)
```

```python
data = {'x_1': [], 'x_2': [], 'y': []}
for i in range(500):
    data['x_1'].append(x_1[i])
    data['x_2'].append(x_2[i])
    data['y'].append(sin_sum(x_1[i], x_2[i]))
```

```python
out_df = pd.DataFrame(data=data)
out_df
```

```python
out_df.to_csv('y=sin(x_1 + x_2).csv', index = False)
```

## 2

```python
df = pd.read_csv('y=sin(x_1 + x_2).csv')
df
```

```python
x_2_const = df.x_2[0]
y_fun_of_x_1 = [sin_sum(x_1, x_2_const) for x_1 in df.x_1]
plt.plot(df.x_1, y_fun_of_x_1, marker='o')
plt.xlabel('x_1', size=20)
plt.ylabel('y', size=20)
plt.title('Plot for $y=sin(x_1 + x_2)$ where $x_2$ = const', size=20)


plt.show()
```

```python
x_1_const = df.x_1[0]
y_fun_of_x_2 = [sin_sum(x_1_const, x_2) for x_2 in df.x_2]
plt.plot(df.x_2, y_fun_of_x_2, marker='x')
plt.xlabel('x_2', size=20)
plt.ylabel('y', size=20)
plt.title('Plot for $y=sin(x_1 + x_2)$ where $x_1$ = const', size=20)


plt.show()
```

## 3 find mean, min, max of $x_1, x_2, y$

```python
x_1_mean = df.x_1.mean()
x_2_mean = df.x_2.mean()
y_mean = df.y.mean()
print(f'x_1 mean is {x_1_mean}\nx_2 mean is {x_2_mean}\ny mean is {y_mean}\n')

x_1_min = df.x_1.min()
x_2_min = df.x_2.min()
y_min = df.y.min()
print(f'x_1 min is {x_1_min}\nx_2 min is {x_2_min}\ny min is {y_min}\n')


x_1_max = df.x_1.max()
x_2_max = df.x_2.max()
y_max = df.y.max()
print(f'x_1 max is {x_1_max}\nx_2 max is {x_2_max}\ny max is {y_max}\n')

```

## 4 $x_1$ < $x_1 mean$ or $x_2$ < $x_2 mean$

```python
condition_values= df[(df.x_1 < x_1_mean) | (df.x_2 < x_2_mean)]
condition_values
```

```python
condition_values.to_csv('condition_values.csv', index=False)
```
